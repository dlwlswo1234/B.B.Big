```python
article_list = json.loads(html)["news"]["resultList"]
```

```python
titles = []
contents = []
for i in range(0,len(article_list)):
    titles.append(article_list[i]["TITLE"])
    contents.append(article_list[i]["CONTENT"])
    
```

```python
from konlpy.tag import Twitter 
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
tagger = Twitter() 
```

```python
def kor_noun(text):
    words = []
    for w in tagger.morphs(text):
        if len(w) > 1:
            words.append(w)
        return words 
```

```python
cv = CountVectorizer(tokenizer=kor_noun, max_features=1000) 
tdm = cv.fit_transform(titles)
words = cv.get_feature_names() 
```

```python
sorted(zip(tdm.sum(axis=0).flat, words), reverse=True)[:100] 
```

```python
from gensim.matutils import Sparse2Corpus
```

```python
corpus = Sparse2Corpus(tdm.T) 
```

```python
from gensim.models.ldamodel import LdaModel

num_topics=3

lda = LdaModel(corpus=corpus, id2word=dict(enumerate(words)), 
               num_topics=3, passes=10, iterations=50, random_state=1234) 
```

```python
for t in range(num_topics):
    topic = [word for word, p in lda.show_topic(t)]
    print(t, ' '.join(topic)) 
```
